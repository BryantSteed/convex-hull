```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from convex_hull import compute_hull_dvcq
from main import main
from plotting import draw_hull, show_plot, title,  plot_points
```

```python
first_date = "02/27/2020"
second_date = "03/12/2020"
third_date = "03/27/2020"

with open("Covid_19_Countrywise_timeseries.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)  # Skip header row

    points_first = []
    points_second = []
    points_third = []

    number_logged = 0

    for row in reader:

        date = row[1]

        try:
            lat = float(row[7])
            lon = float(row[8])
        except ValueError:
            continue  # Skip rows with invalid lat/lon

        if date == first_date:
            points_first.append((lon, lat))
            number_logged += 1
        elif date == second_date:
            points_second.append((lon, lat))
            number_logged += 1
        elif date == third_date:
            points_third.append((lon, lat))
            number_logged += 1
    
    print(f"Total number of logged points: {number_logged}")

        
```

```python
hull_points1 = compute_hull_dvcq(points_first)
plot_points(points_first)
draw_hull(hull_points1)
title(f'Points for {first_date} of Covid-19')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
show_plot()

hull_points2 = compute_hull_dvcq(points_second)
plot_points(points_second)
draw_hull(hull_points2)
title(f'Points for {second_date} of Covid-19')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
show_plot()

hull_points3 = compute_hull_dvcq(points_third)
plot_points(points_third)
draw_hull(hull_points3)
title(f'Points for {third_date} of Covid-19')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
show_plot()

print(len(hull_points1))
print(len(hull_points2))
print(len(hull_points3))
```
