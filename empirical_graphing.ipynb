```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 100, 1000, 10000, 20000, 40000, 50000])
y = np.array([0.1, 0.8, 8.3, 80.3, 181.6, 335.4, 448.1])

nlogn = x * np.log2(x)
nlogn_constants = y / nlogn
average_logn_constant = np.mean(nlogn_constants)

linear = x
linear_constants = y / linear
average_linear_constant = np.mean(linear_constants)
```

```python
# Bar plot of constants (unsorted)
constants = nlogn_constants  # uses values from previous cell
idx = np.arange(constants.size)

plt.figure(figsize=(8, 4))
plt.bar(idx, constants, color="#4C78A8")
plt.axhline(average_logn_constant, color="crimson", linestyle="--", label=f"avg = {average_logn_constant:.3g}")
plt.xlabel("Sample index")
plt.ylabel("Constant (time / (n log2 n))")
plt.title("Constants for n log n time complexity")
plt.xticks(idx)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Bar plot of linear-time constants (unsorted)
constants = linear_constants
idx = np.arange(constants.size)

plt.figure(figsize=(8, 4))
plt.bar(idx, constants, color="#72B7B2")
plt.axhline(average_linear_constant, color="crimson", linestyle="--", label=f"avg = {average_linear_constant:.3g}")
plt.xlabel("Sample index")
plt.ylabel("Constant (time / n)")
plt.title("Constants for linear time complexity")
plt.xticks(idx)
plt.legend()
plt.tight_layout()
plt.show()

plt.clf()
```

```python
plt.clf()
plt.plot(x, y, marker='o', label='Empirical Data')
plt.plot(x, average_logn_constant * nlogn, label='n log n (theoretical) Fit', linestyle='--')
plt.plot(x, average_linear_constant * linear, label='Linear Fit', linestyle='--')
plt.xlabel("Input Size (n)")
plt.ylabel("Time (ms)")
plt.legend()
plt.show()
```
